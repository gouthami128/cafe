```python
import pandas as pd
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
data = pd.read_csv("/amazon_reviews - amazon_reviews.csv")
data.head()
```

```python
s = sia.polarity_scores(data.iloc[0]['label\treview'])
s
```

```python
s['compound']
```

```python
score = []
for i in range(len(data)):
  s = sia.polarity_scores(data.iloc[i]['label\treview'])
  score.append(s)
```

```python
data['Scores'] = score
```

```python
data.head()
```

```python
comp = []
for i in range(len(data)):
  s = sia.polarity_scores(data.iloc[i]['label\treview'])
  comp.append(s['compound'])
```

```python
data['Compound'] = comp
data.head()
```

```python
sentiments = []
for i in range(len(data)):
  s = sia.polarity_scores(data.iloc[i]['label\treview'])
  if s['compound'] > 0:
    sentiments.append("pos")
  else:
    sentiments.append("neg")
```

```python
data['Sentiment'] = sentiments
data.head()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(data['label\treview'],data['Sentiment'])
cm
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(data['label\treview'],data['Sentiment'])
a
```
